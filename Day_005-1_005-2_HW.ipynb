```python
import requests
import pandas as pd

url = 'https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'
try:
    htmlfile = requests.get(url)
    print('â∫ç⁄ê¨å˜')
except Exception as err:
    print('â∫ç⁄é∏îs')
```

```python
#õíèäíÒãüìI txt ë∂ê¨ imagenet_urls_examples.csv
fn = 'imagenet_urls_examples.csv'
with open(fn,'wb') as fileobj:
    for data in htmlfile.iter_content(10240):
        size = fileobj.write(data)
```

```python
colName = ['filename','url']
#pd.set_option('display.max_colwidth',70)
df = pd.read_csv(fn,names = colName,sep = '\t')
print (type(df))
print (df) #pandas dataframe
```

```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from PIL import Image

df_to_nda = np.array(df) #np.ndarray()
df_to_list = df_to_nda.tolist() #list

i = 0

#ëO5í£ñ‘ö¨ñ¢é∏ù¡ìIö§ï–
for img in df_to_list:
    img_src = img[1]
    try:
        image = io.imread(img_src)
    except Exception as err:
        print('URL Not Found:', img[1])
        continue

    plt.axis("off")
    plt.imshow(image)
    plt.show()
        
    i=i+1
    
    if i ==5:
        break
```
