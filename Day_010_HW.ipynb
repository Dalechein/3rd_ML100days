```python
# 做完特徵工程前的所有準備 
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

df_train = pd.read_csv('house_train.csv')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
df['1stFlrSF'].describe()
```

```python
# 調整離群值(箱型法)
iqr = df['1stFlrSF'].std() 
u_th = df['1stFlrSF'].quantile(0.75) + 1.5 * iqr
l_th = df['1stFlrSF'].quantile(0.25) - 1.5 * iqr

df2 = df.replace(df[df['1stFlrSF'] < l_th], l_th)
df2 = df.replace(df[df['1stFlrSF'] > u_th], u_th)

sns.regplot(x = df2['1stFlrSF'], y = train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df2)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 調整離群值(Z分數法)
u_th = df['1stFlrSF'].mean() + 3 * iqr
l_th = df['1stFlrSF'].mean() - 3 * iqr

df3 = df.replace(df[df['1stFlrSF'] < l_th], l_th)
df3 = df.replace(df[df['1stFlrSF'] > u_th], u_th)

sns.regplot(x = df3['1stFlrSF'], y = train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df3)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF 限制在適合的範圍內, 捨棄離群值
#u_th = df['1stFlrSF'].quantile(0.975)
#l_th = df['1stFlrSF'].quantile(0.025)
keep_indexs =  (df['1stFlrSF'] > l_th) & (df['1stFlrSF'] < u_th)
df = df[keep_indexs]
train_Y = train_Y[keep_indexs]
sns.regplot(x = df['1stFlrSF'], y = train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

'捨棄離群值' 的結果比'調整離群值'的結果好, 因為資料中存在極端值, 影響了標準差和平均值, 捨棄離群值後, 結果變好。
